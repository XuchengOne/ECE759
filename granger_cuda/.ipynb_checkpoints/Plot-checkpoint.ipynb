```python
# libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = [i for i in range(1,7)]
y = [1.02283,
   1.02291,
   0.521404,
   0.526444,
   0.339829, 
   1.01525]
```

```python
# Data
df=pd.DataFrame({'Optimize X': x, 'Time': y})
 
# multiple line plot
fig, ax = plt.subplots(figsize=(10,6), dpi=150)
ax.plot( 'Optimize X', 'Time', data=df, marker='o',linestyle='dashed')

plt.title("Task 14")
plt.xlabel("Optimize X")
plt.ylabel("Time(ms)")

plt.legend()
plt.savefig("task14.pdf", bbox_inches='tight')
```
